# FER-2013 — Exploratory Data Analysis
Quick look at the dataset before training.

```python
import sys
sys.path.insert(0, '..')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from utils import load_fer2013, plot_class_distribution

%matplotlib inline
plt.rcParams['figure.dpi'] = 120
```

```python
df = pd.read_csv('../data/fer2013.csv')
df.head(3)
```

```python
plot_class_distribution(df['emotion'], title='FER-2013 full dataset class distribution')
```

```python
# Show 5 random samples per emotion
EMOTION_LABELS = ['Angry','Disgust','Fear','Happy','Neutral','Sad','Surprise']
df['pixels_arr'] = df['pixels'].apply(lambda s: np.array(s.split(), dtype=np.float32).reshape(48,48))

fig, axes = plt.subplots(7, 5, figsize=(10, 14))
for emo_idx, lbl in enumerate(EMOTION_LABELS):
    samples = df[df['emotion'] == emo_idx].sample(5, random_state=42)
    for col, (_, row) in enumerate(samples.iterrows()):
        axes[emo_idx, col].imshow(row['pixels_arr'], cmap='gray')
        axes[emo_idx, col].axis('off')
        if col == 0:
            axes[emo_idx, col].set_title(lbl, fontsize=9, loc='left')
plt.suptitle('5 samples per emotion class', y=1.01, fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Pixel intensity statistics
pixels = np.stack(df['pixels_arr'].values)
print(f'Mean pixel value : {pixels.mean():.2f}')
print(f'Std  pixel value : {pixels.std():.2f}')
print(f'Min              : {pixels.min()}')
print(f'Max              : {pixels.max()}')

plt.hist(pixels.flatten(), bins=50, color='steelblue', edgecolor='white')
plt.title('Pixel intensity distribution (all images)')
plt.xlabel('Pixel value (0–255)')
plt.ylabel('Count')
plt.show()
```
